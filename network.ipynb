```python
import networkx as nx
import matplotlib.pyplot as plt
import string
import pandas as pd
```

```python
LL=pd.read_csv("LL.csv")
LF=pd.read_csv("LF.csv")
LL=LL.drop("Unnamed: 0",axis=1)
LF=LF.drop("Unnamed: 0",axis=1)
LL=LL.rename(columns={"0": 'ID'})
```

```python
LF.loc[0][0].astype(int)
LG=pd.concat([LL,LF],axis=1)
#LG.shape,LL.shape,LF.shape
LG.loc[0][0].astype(int)
LG=LG.fillna(0).astype(int)
```

```python
LF.shape,LL.shape
#削除命令
#G.remove_node(LL['ID'][3])
G=nx.DiGraph()
```

```python
#このシェルはネットワークデータ定義用
"""本番用コード
for i in LL["ID"]:
    #G.add_node(LL['ID'][i])
    G.add_node(i)
以下練習用小規模VER.
"""
for i in range(5):
    G.add_node(LG.loc[i][0].astype(int))
    for j in range(len(LG.loc[i].dropna().astype(int)[1:])-1):
        G.add_edge(LG.loc[i][0].astype(int),LG.loc[i].dropna().astype(int)[1:][j])
```

```python
#図の作成。figsizeは図の大きさ
plt.figure(figsize=(10, 8))
 
#図のレイアウトを決める。kの値が小さい程図が密集する
pos = nx.spring_layout(G, k=0.8)
#ノードとエッジの描画
# _color: 色の指定
# alpha: 透明度の指定
nx.draw_networkx_edges(G, pos, edge_color='y')
nx.draw_networkx_nodes(G, pos, node_color='r', alpha=0.5)
#ノード名を付加
nx.draw_networkx_labels(G, pos, font_size=7)
 
#X軸Y軸を表示しない設定
plt.axis('on')
 
#図を描画
plt.show()
```

```python
plt.figure(figsize=(7, 7))
pos = nx.spring_layout(G, k=0.7)
#PageRankの追加
pr = nx.pagerank(G)
nx.draw_networkx_edges(G, pos, edge_color='y')
#node_sizeにPageRankの値を組み込む
nx.draw_networkx_nodes(G, pos, node_color='r', alpha=0.5, node_size=[5000*v for v in pr.values()])
nx.draw_networkx_labels(G, pos, font_size=10)
plt.axis('on')
plt.show()
```
